```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("C:/Users/saads/Downloads/train.csv/train.csv")
```

```python
data.head(20)
```

```python
#converting the csv into array
data = np.array(data)
#then we randomly shuffle the data just for the sake of it
np.random.shuffle(data)
# here we initalise that the  shape of the data is what reference to m and n variable repectively
m, n = data.shape
# m = 420000 samples of data, n = the length of the input layer 

#we drag 1000 rows form the data of 42000 rows as the dev data which we use to test .. as this data is completely unseen for the model
# acts as a real world implementation
data_dev = data[0:1000].T 

#this takes the first layer as the target
Y_dev = data_dev[:, 0]
#and the rest goes to the X_dev
X_dev = data_dev[:, 1:]

#now we train the data on the rest of remaining data set
data_train = data[1000:m].T
#here we are intializing the y for the first column of the db that has the labels
Y_train = data_train[0]
#and rest to the x_train which contins the each pixel of the image with pixel value
X_train = data_train[1:n] # n being the number of pixel which is 784 also called the number of row
```

```python
data.shape
```

```python
#now the reason why we are doing this is becasue the pixel white color is denoted as 255 and grey as 219
#now image has 784 such pixel that means if we apply any sort of computation the number is goinf to be crazy big hence we bring
#down the values to 1 for white and less for the less white(grey, light grey, dark grey, black)
X_train = X_train / 255.0
#same for the dev 
X_dev = X_dev / 255.0
```

```python
#initializing parameters
#we do this step because our model change the weights and bias but it need a place to start with hence we provide it by 
#intializing it 

def init_params():
    #so here we are using radn which generates value between 0 and 1 for (x,y) matrices here x and y are 10 , 784 .. 
    #784 neuron connecting with 10 nueron in 1st hidden .... 10x784 ... hence we use randn to generate randome value of weights
    #between 0 and 1 and we use np.sqrt to scale the varaince so that the varaince is not explodig or to small
    W1 = np.random.randn(10, 784) * np.sqrt(2 / 784)
    #for bias we use 10 x 1 matric cause only one bias per 1st hidden layer and we dont need sqrt as it never explodes
    b1 = np.zeros((10, 1))
    W2 = np.random.randn(10, 10) * np.sqrt(2 / 10)
    b2 = np.zeros((10, 1))

    return W1,b1,W2,b2


#here we are implementing ReLU and as we know it is based on the principle of ...if the value is less than 0 it will be zero 
#else it will reflect what ever value is getting reflected
# we use ReLU for 2 hidden layer only
def ReLU(Z):
    return np.maximum(0,Z) #this take 0 if z is negative or takes z if z is positive 


#this is another activation funciton used when the output is multi class and return single value which is the biggest value
def softmax(Z):
    #sometime z is crazy big due to which exp become infinte number hence Nan, so we deduct eveery single z with a maximum value 
    # kinda like with the same constant so the z is same but very small( 10001, 10003, 9999 -> 1,3,-2)
    Z = Z - np.max(Z, axis=0, keepdims=True)  
    expZ = np.exp(Z)
    #using axis is you ensure that we are applying computing at one same axis and keepdim is just for formating
    return expZ / np.sum(expZ, axis=0, keepdims=True)  # sum per sample

#this is normal forward propagation
def forward_prop(W1,b1,W2,b2,X):
    Z1 = W1.dot(X) + b1
    #we apply relu activation for 1st hidden layer
    A1 = ReLU(Z1)
    Z2 = W2.dot(A1) + b2
    # but appply softmax for the output layer as our output is multiclass.
    A2 = softmax(Z2)
    return Z1,A1,Z2,A2

#this is like converting the a label data into sparse data 

def one_hot(Y):
    
    #so here we are creating a matric of zeroes with dimention of (y.size,y.max) (10,9+1)
    one_hot_Y = np.zeros((Y.size, Y.max()+1))
    #here we are assigning values to every indinces the lbel is crossing at
    #so one_hot_y [10,0]=1 1o5h row 0th column will have 1 as the label regarldes off what the values is 
    one_hot_Y[np.arange(Y.size),Y] = 1
    #then we transpose the matrix
    one_hot_Y = one_hot_Y.T
    return one_hot_Y

#this is use to find the deviation of the activation funciton
# which is if the z is postive .. it is the deviation is not then its 0
def deriv_ReLU(Z):
    return Z > 0
    
#here we are implenting the first and most important part of the whole NN core
#backpropogation simple means the re iteration of th forward propagation and make changes to get the desired result
def backpropagation(Z1,A1,Z2,A2,W2,X,Y):
     #m size is of y as we are in the hidden layer right nwo and our output is 10 posssible output class
    m = Y.size
    # we use our one_hot funciton with y label
    one_hot_Y = one_hot(Y)
    #now we progress backward as we look at the last encoding and subtract with the activation value toc heck how much is the 
    #differnce also caled the deviation we need so for the z2 its dz2
    dZ2 = A2 - one_hot_Y
    #thats how we calculate teh weight deviation 
    dW2 = 1 / m * dZ2.dot(A1.T)
    db2 = 1 / m * np.sum(dZ2, 0)
    #here we are calcualting the deviation of the z1 by applyig the mathematical formula of w2xdz2 * activation of the z1
    dZ1 = W2.T.dot(dZ2) * deriv_ReLU(Z1)
    #again calculate weight and biases for the first hidden layer
    dW1 = 1 / m * dZ1.dot(X.T)
    db1 = 1 / m * np.sum(dZ1, 0)
    return dW1, db1, dW2, db2

#here we are implementing the 2nd most important process of deep learning
#the ability to change the parameters and adjust to get the desired predicited result 
#it takes the inital weight and baises, the deviated weight and biases and the learning rate
def update_parameters(W1,b1,W2,b2,dW1,db1,dW2,db2,alpha):
    #so we do this to update the weight and biases 
    #we subtract the deviation * alpha result as our agenda is to move towards the small gradient for whicch we need to reduce the weight 
   #hence reducing the weight and biases
    W1 = W1 - alpha * dW1
    b1 = b1 - alpha * db1
    W2 = W2 - alpha * dW2
    b2 = b2 - alpha * db2
    return W1, b1, W2, b2
```

```python
#this jsut return of the activation result from the output layer if its bigger thn 0 as it need to be a number positvie
def get_predictions(A2):
    return np.argmax(A2,0)

#here we are implementing the accuracy of the model
def get_accuracy(predictions, Y):
    print(predictions, Y)
    #here we check y^ with y to check the acuracy and to keep the accruacy in same ballpark 0,1 hence we devide it by 10 or the y size
    return np.sum(predictions == Y) / Y.size

train_accuracy_list = []

#here we implenting the vanilla sgd and describing how we are going progress in this method
def gradient_descent(X, Y, iterations, alpha):
    #first step initalize the randome weights
    #this happen outside the loop as we need to happen only the first time as later the algo figure out its own
    W1, b1, W2, b2 = init_params()
    #then we do the iteration and make sure dont iterate to much as it would lead to overfitting
    for i in range(iterations):
        #here we find out z1,a1,z2,a2 from the forward_prop method
        Z1, A1, Z2, A2 = forward_prop(W1, b1, W2, b2 , X)
        #here we find out the deviation using the backpropogation
        dW1, db1, dW2, db2 = backpropagation(Z1,A1,Z2,A2,W2,X,Y) 
        
        #now after finding ho much we need to change we update the parameters using this method
        W1, b1, W2, b2 = update_parameters(W1, b1, W2, b2, dW1, db1, dW2, db2, alpha)
        
        #now we print everytime the model iterate the number of time the iterating is set up to also check the  gradual increase in the
        #prediction accuracy as it keep checking with the label and show us a real time gradient result
        if(i % 5 == 0):
            acc = get_accuracy(get_predictions(A2), Y)
            train_accuracy_list.append(acc)
            print(f"Iteration {i}, Accuracy: {acc:.4f}")
    return W1,b1,W2,b2
  
```

```python
#here we are applying the sgd with x and  y as the train value with 500 iterations and 0.1 learing rate hyper parameter something we 
#initlize and nto something the algo learn
W1, b1, W2, b2 = gradient_descent(X_train, Y_train, 500 , 0.1)

#and now we can see our model learning in real time as the accuracy number gradually increases and at the end we see after 500 
#iteration it reaches 90 percent accuracy which is amazing

# xxx-----The End------xxx
```

```python
def make_predictions(X,W1,b1,W2,b2):
    _, _, _, A2 = forward_prop(W1, b1, W2, b2, X)
    predictions = get_predictions(A2)
    return predictions

def test_prediction(index, W1, b1, W2, b2):
    current_image = X_train[:, index, None]
    prediction = make_predictions(X_train[:, index, None], W1, b1, W2, b2)
    label=Y_train[index]
    print("Prediction: ",prediction)
    print("Label:, ",label)
          
    current_image=current_image.reshape((28, 28))*255
    plt.gray()
    plt.imshow(current_image, interpolation= "nearest")
    plt.show()
```

```python
test_prediction(135,W1,b1,W2,b2)
```

```python
#plotting the accuracy plot of our model
epochs = list(range(0, len(train_accuracy_list)*5, 5)) 
plt.figure(figsize=(6,4))
plt.plot(epochs, train_accuracy_list, marker='o', label="Training Accuracy")
plt.xlabel("Iterations")
plt.ylabel("Accuracy")
plt.title("Training Accuracy over Iterations")
plt.legend()
plt.grid(True)
plt.show()
```
